```python
import matplotlib.pyplot as plt
import numpy as np

# 2行3列のサブプロットを作成
fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

# sin関数
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)

# ヒストグラム
data = np.random.randn(1000) * 10 + 50
axes[0, 2].hist(data, bins=20)

# 散布図
x2 = np.random.rand(50) * 10
y2 = np.random.rand(50) * 10
axes[1, 0].scatter(x2, y2)

# 空のプロット（そのままでもOK）
axes[1, 1].set_xticks([])
axes[1, 1].set_yticks([])
axes[1, 2].set_xticks([])
axes[1, 2].set_yticks([])

# グラフを表示
plt.show()
```
