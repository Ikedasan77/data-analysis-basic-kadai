```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# CSVファイルのパス
file_path = "/content/sample_pandas_6.csv"

df = pd.read_csv(file_path)

category_counts = df['商品名'].value_counts()
category_counts.plot(kind='bar')

# Step3: グラフを表示
plt.xlabel("商品名")
plt.ylabel("出現頻度")
plt.title("商品ごとの出現頻度")
plt.xticks(rotation=90)  # ラベルが見やすいように回転
plt.show()

# Step4: 商品番号ごとの注文数の基本統計量を算出
order_stats = df.groupby('商品番号')['注文数'].describe()

# 結果を表示（ace_tools の代わりに print を使用）
print(order_stats)
```
